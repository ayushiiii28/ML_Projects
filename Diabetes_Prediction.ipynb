```python
#Importing Dependencies
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler #for standardizing the data to a common range
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
```

```python
#Data collection and analysis
#Uploading dataset
diabetes_dataset = pd.read_csv('Downloads/diabetes.csv')
df= pd.DataFrame(diabetes_dataset)
print(df)
```

```python
#printing 1st 5 rows
df.head()
```

```python
#rows and columns
df.shape
```

```python
#Getting the statistical measures of the data
df.describe()
```

```python
df['Outcome'].value_counts()
```

```python
df.groupby('Outcome').mean()
```

```python
#separating the data and labels
x= df.drop(columns='Outcome', axis=1)
y= df['Outcome']
```

```python
print(x)
```

```python
print(y)
```

```python
y.shape
```

```python
#Data Standardization
scaler= StandardScaler()
```

```python
scaler.fit(x)
```

```python
standardized_data = scaler.transform(x)
```

```python
print(standardized_data)
```

```python
x=standardized_data
y= df['Outcome']
```

```python
print(x)
print(y)
```

```python
#Train Test Split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, stratify=y, random_state=2)
#0.2 represents 20% of test data and rest is training data
#stratify=y, where y is vallues 0,1, for data set to be splitted in same proportion and to avoid misplacing of data
#random_state is to spliting of data, acc to code
#x,y are attributes
```

```python
print(x.shape,x_train.shape,x_test.shape)
```

```python
#Training the model
classifier = svm.SVC(kernel='linear')
```

```python
#fitting training data to svm classifier
classifier.fit(x_train,y_train)
```

```python
#Model Evaluation
#Accuracy Score on training data
x_train_prediction = classifier.predict(x_train)
training_data_accuracy = accuracy_score(x_train_prediction, y_train)
```

```python
print("Accuracy score of training score: ", training_data_accuracy)
```

```python
# accuracy Predictive datascore on the test data
x_test_prediction = classifier.predict(x_test)
test_data_accuracy = accuracy_score(x_test_prediction, y_test)
```

```python
print(f"Accuracy score of the test data : {test_data_accuracy}")
```
